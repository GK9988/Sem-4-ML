```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/DataPreprocessing.csv")
df.head()
```

```python
np.round(df['MSSubClass'])
np.round(df['LotArea'])
Mean=df['MSSubClass'].mean()
print(Mean)
stand=df['MSSubClass'].std()
print(stand)
mean_2=df['LotArea'].mean()
print(mean_2)
stand_2=df['LotArea'].std()
print(stand_2)
```

```python
df['LotArea']=df['LotArea'].apply(lambda x:((x-mean_2)/stand_2))
```

```python
df['LotArea']
```

```python
df['LotArea'].describe()
```

```python
df['MSSubClass']=df['MSSubClass'].apply(lambda x:((x-Mean)/stand))
```

```python
df['MSSubClass'].describe()
```

```python
df.replace([':-','?','_','-','#','##','Na','Nan'],np.nan,inplace=True)
```
