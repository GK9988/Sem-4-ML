```python
import pandas as pd
import numpy as np
df = pd.read_csv('DataPreprocessing.csv')
df
```

```python
mean = np.round(df['MSSubClass'].mean())
mean
```

```python
std = df['MSSubClass'].std()
std
```

```python
df['MSSubClass'] = df['MSSubClass'].apply(lambda x: ((x-mean)/std))
df['MSSubClass'].describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df['LotArea']=scaler.fit_transform(df[['LotArea']])
df['LotArea']
```

```python
from sklearn.preprocessing import MaxAbsScaler
max_abs=MaxAbsScaler()
```

```python
df['TotalBsmtSF']=max_abs.fit_transform(df[['TotalBsmtSF']])
df['TotalBsmtSF']
```

```python
a=df['TotalBsmtSF'].max()
df['TotalBsmtSF']=df['TotalBsmtSF'].apply(lambda x:((x)/abs(a)))
df['TotalBsmtSF'].describe()
```

```python
a=df['GrLivArea'].max()
b=df['GrLivArea'].min()
df['GrLivArea']=df['GrLivArea'].apply(lambda x:((x-b)/(a-b)))
df['GrLivArea'].describe()
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
df
```

```python
scalar = StandardScaler()
```

```python
scaled_data = scalar.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(scaled_data)
x_pca.shape
```
